We know by the previous analysis, that there are way more leeches in the Englisch Wortschatz - deck, than in the daily knowledge deck. Let us take a look how the distribution of the "ease-factor" differ. 

We start with some data cleaning !

```python
import seaborn as sns
import pandas as pd

cards_all_data=pd.read_csv("cards_all_data.csv")

cards_english=cards_all_data[cards_all_data["cdeck"]=="Englisch Wortschatz"].copy()
cards_daily_know=cards_all_data[cards_all_data["cdeck"]=="daily knowledge"].copy()

cards_english=cards_english[cards_english["cfactor"]!=0]
cards_daily_know=cards_daily_know[cards_daily_know["cfactor"]!=0]
all_cards=pd.concat([cards_english,cards_daily_know])
sns.displot(all_cards, x="cfactor", hue="cdeck",kind="kde",fill=True,bw_adjust= 1.5)
# sns.displot(cards_english["cfactor"],x="ease factor, Englisch Wortschatz deck")
```
